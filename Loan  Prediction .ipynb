```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split 
from sklearn.ensemble import RandomForestClassifier
```

```python
df=pd.read_csv("C:/Users/91976/Documents/MSC S2/ML/ML Project 11/ML Project 1 Dataset.csv")
df
```

```python
df.info()
```

```python
df.columns.values
```

```python
df['NEW_CUST'].replace({'NO': 0, 'YES': 1},inplace = True)
df[' SEX  '].replace({'F': 0, 'M': 1},inplace = True)
df.head()
```

```python
df.shape
```

```python
x=df.drop(columns=['APP_ID','STATUS'], axis=1)
y=df['STATUS']
x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,stratify=y,random_state=0)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
print(y.shape,y_train.shape,y_test.shape)
```

```python
forest = RandomForestClassifier(n_estimators=10,max_depth=3, criterion='entropy')
forest.fit(x_train, y_train)
```

```python
x_train_prediction=forest.predict(x_train)
training_data_accuracy=accuracy_score(x_train_prediction,y_train)
training_data_accuracy
```

```python
x_test_prediction=forest.predict(x_test)
testing_data_accuracy=accuracy_score(x_test_prediction,y_test)
testing_data_accuracy
```
